```
from google.colab.patches import cv2_imshow
```

```
#LECTURA DE IMAGEN
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```
img1 = cv2.imread('/content/img1_CLAHE.png',0)
```

```
cv2_imshow(img1)
```

```
# GAUSSIAN BLUR
cl1 = cv2.GaussianBlur(img1,(3,3),0)
cv2_imshow(cl1)
```

```
# BLUR
cl2 = cv2.blur(cl1,(3,3))
cv2_imshow(cl2)
```

```
# BLUR
cl3 = cv2.medianBlur(cl2,3)
cv2_imshow(cl3)
```
